```python

# ---------------------------------------------------------------------
# 1. Load your filtered FXRT candidate table
# ---------------------------------------------------------------------
# Expect columns:  RA  DEC  pos_err  (all from new_fxrt_w0.2_possible.csv)
# RA, DEC in degrees; pos_err in arcsec.
df = pd.read_csv("output/candidates/new_fxrt_w0.2_possible.csv")

print(f"Loaded {len(df)} candidates with columns:", df.columns.tolist())
df.head()

# ---------------------------------------------------------------------
# 2. Boresight (WCS) translation offsets
# ---------------------------------------------------------------------
# If you already derived mean offsets (ΔRA*cosDec, ΔDec) from wcs_match residuals, set them here (arcsec):
dx_mean_arcsec = 0.0   # e.g. +0.20 arcsec east  → positive RA
dy_mean_arcsec = 0.0   # e.g. −0.15 arcsec south → negative Dec

# Apply translation to all coordinates
df["RA_CORR"]  = df["RA"]  + dx_mean_arcsec/3600.0 / np.cos(np.deg2rad(df["DEC"]))
df["DEC_CORR"] = df["DEC"] + dy_mean_arcsec/3600.0

# ---------------------------------------------------------------------
# 3. Determine what your pos_err represents
# ---------------------------------------------------------------------
# Your filter script's pos_err is almost certainly a 1-σ radial statistical error.
# If you're confident it is r90 or r95 instead, set MODE accordingly.
MODE = "sigma"   # options: "sigma", "r90", "r95"

if MODE == "sigma":
    sigma_stat = df["POS_ERR"].astype(float).values
elif MODE == "r90":
    sigma_stat = df["POS_ERR"].astype(float).values / 2.146
elif MODE == "r95":
    sigma_stat = df["POS_ERR"].astype(float).values / 2.448
else:
    raise ValueError("MODE must be 'sigma', 'r90', or 'r95'.")

# ---------------------------------------------------------------------
# 4. Add systematic alignment RMS (boresight uncertainty)
# ---------------------------------------------------------------------
# If you computed a residual RMS from wcs_match residuals, set it here (arcsec).
# Otherwise leave it 0.0 — you can re-run this cell later once known.
sigma_sys = 0.0    # arcsec
sigma_tot = np.sqrt(sigma_stat**2 + sigma_sys**2)

# ---------------------------------------------------------------------
# 5. Compute 95% confidence radius (circular)
# ---------------------------------------------------------------------
k95 = np.sqrt(5.991)        # = 2.448, for 2-D Gaussian 95% containment
R95_arcsec = k95 * sigma_tot
df["R95_arcsec"] = R95_arcsec

print("Typical 95% error radius (arcsec):",
      np.nanmedian(df["R95_arcsec"]), "median")

# ---------------------------------------------------------------------
# 6. (Optional) Self-calibrate scale factor using control matches
# ---------------------------------------------------------------------
# If you have a control set (e.g. bright X-ray ↔ Gaia matches) you can refine scaling.
# Uncomment and fill in if available.
"""
ctrl = pd.read_csv("control_matches.csv")  # columns: sep_arcsec, pos_err
s_grid = np.linspace(0.6, 1.6, 101)
contain = []
for s in s_grid:
    r95_ctrl = 2.448 * s * ctrl["pos_err"].values
    frac = np.mean(ctrl["sep_arcsec"].values <= r95_ctrl)
    contain.append(frac)
s_best = s_grid[np.argmin(np.abs(np.array(contain) - 0.95))]
print(f"Empirical scale factor s ≈ {s_best:.2f}")
R95_arcsec = 2.448 * np.sqrt((s_best*sigma_stat)**2 + sigma_sys**2)
df["R95_arcsec"] = R95_arcsec
"""

# ---------------------------------------------------------------------
# 7. Save the results
# ---------------------------------------------------------------------
if "ObsId" not in df.columns:
    raise KeyError("Column 'ObsId' not found in input CSV. Please check the column name.")

out_cols = ["ObsId", "RA_CORR", "DEC_CORR", "R95_arcsec"]

df[out_cols].to_csv(
    "crossmatches/fxrt_positions_R95.txt",
    sep="\t",
    index=False,
    float_format="%.6f"
)

print(f"✅ Saved {len(df)} corrected positions to fxrt_positions_R95.txt")
print(df[out_cols].head())

# ---------------------------------------------------------------------
# 8. Quick sanity check plot (optional)
# ---------------------------------------------------------------------


plt.hist(df["R95_arcsec"], bins=30, histtype='step', color='navy')
plt.xlabel("95% positional radius (arcsec)")
plt.ylabel("Number of FXRT candidates")
plt.title("Distribution of R95 positional uncertainties")
plt.show()
```

```python
# ------------------------------------------------------------------
# 1.Optical Matches Pan-STARRS DR2
# ------------------------------------------------------------------
# Expected columns: ObsID, RA_CORR, DEC_CORR, R95_arcsec (in arcsec)
fxrt = pd.read_csv("crossmatches/fxrt_positions_R95.txt", sep="\t")

print(f"Loaded {len(fxrt)} FXRT candidates with columns:", fxrt.columns.tolist())
fxrt.head()

# ------------------------------------------------------------------
# 2. Define Pan-STARRS DR2 query function
# ------------------------------------------------------------------
def query_ps1(ra, dec, rad_arcsec=2.0):
    """
    Query Pan-STARRS DR2 mean catalog for one position using MAST API.
    Returns a dictionary with RA, Dec, and gMeanPSFMag.
    """
    base = "https://catalogs.mast.stsci.edu/api/v0.1/panstarrs/dr2/mean.csv"
    rad_deg = rad_arcsec / 3600.0
    query = f"{base}?ra={ra:.6f}&dec={dec:.6f}&radius={rad_deg:.6f}"

    # Default output in case no match found
    out = {'ra_ps1': np.nan, 'dec_ps1': np.nan, 'gmag': np.nan, 'sep_arcsec': np.nan}

    try:
        with urllib.request.urlopen(query) as r:
            txt = r.read().decode()

        # Parse CSV lines
        lines = [l for l in txt.splitlines() if l and not l.startswith('#')]
        if len(lines) <= 1:
            return out

        # Read header and first row of data
        header = lines[0].split(',')
        data   = lines[1].split(',')

        # Identify column indices dynamically
        i_ra  = header.index('raMean')
        i_dec = header.index('decMean')
        i_g   = header.index('gMeanPSFMag') if 'gMeanPSFMag' in header else None

        # Fill dictionary
        ra_ps1 = float(data[i_ra])
        dec_ps1 = float(data[i_dec])
        gmag = float(data[i_g]) if i_g is not None and data[i_g] else np.nan

        # Compute angular separation (approx, small-angle)
        dra  = (ra_ps1 - ra) * np.cos(np.deg2rad(dec)) * 3600.0
        ddec = (dec_ps1 - dec) * 3600.0
        sep_arcsec = np.sqrt(dra**2 + ddec**2)

        out.update({'ra_ps1': ra_ps1, 'dec_ps1': dec_ps1, 'gmag': gmag, 'sep_arcsec': sep_arcsec})

    except Exception as e:
        print(f"Pan-STARRS query failed for RA={ra:.3f}, Dec={dec:.3f}: {e}")

    return out


# ------------------------------------------------------------------
# 3. Query Pan-STARRS for all your candidates
# ------------------------------------------------------------------
matches_ps1 = []
for i, row in fxrt.iterrows():
    m = query_ps1(row.RA_CORR, row.DEC_CORR, 1.405*row.R95_arcsec)
    matches_ps1.append(m)
    if i % 10 == 0:
        print(f"Queried {i+1}/{len(fxrt)} sources")

# Convert to DataFrame
ps1_df = pd.DataFrame(matches_ps1)

# Merge with your FXRT table
fxrt_ps1 = pd.concat([fxrt.reset_index(drop=True), ps1_df.reset_index(drop=True)], axis=1)

# ------------------------------------------------------------------
# 4. Save merged output
# ------------------------------------------------------------------
fxrt_ps1.to_csv("crossmatches/fxrt_crossmatch_ps1_3sig.txt", sep="\t", index=False, float_format="%.6f")

print(f"\n✅ Crossmatch complete. Found {np.isfinite(fxrt_ps1['ra_ps1']).sum()} PS1 matches "
      f"out of {len(fxrt_ps1)} candidates.")
print("Saved to fxrt_crossmatch_ps1_3sig.txt")

# ------------------------------------------------------------------
# 5. Optional: view a few results
# ------------------------------------------------------------------
cols = ["ObsId","RA_CORR","DEC_CORR","R95_arcsec","ra_ps1","dec_ps1","gmag","sep_arcsec"]
fxrt_ps1[cols].head(10)
```

```python

# ------------------------------------------------------------------
# WISE / NEOWISE crossmatch  (IRSA Gator) 
# ------------------------------------------------------------------

def query_wise(ra, dec, rad_arcsec=3.0):
    """
    Query the AllWISE Source Catalog for one position via IRSA Gator.
    Returns a dict with WISE RA/Dec, W1, W2 magnitudes, and separation (arcsec).
    """
    base = "https://irsa.ipac.caltech.edu/cgi-bin/Gator/nph-query"
    rad_arcmin = rad_arcsec / 60.0
    params = {
        "catalog": "allwise_p3as_psd",
        "spatial": "Cone",
        "radius": f"{rad_arcmin:.4f}",
        "objstr": f"{ra} {dec}",
        "outfmt": 1,                       # CSV-like text output
        "selcols": "ra,dec,w1mpro,w2mpro"  # select key columns
    }

    url = base + "?" + urllib.parse.urlencode(params)
    out = {'ra_wise': np.nan, 'dec_wise': np.nan,
           'W1mag': np.nan, 'W2mag': np.nan, 'sep_arcsec': np.nan}

    try:
        with urllib.request.urlopen(url, timeout=30) as r:
            text = r.read().decode("utf-8")
        lines = [l for l in text.splitlines()
                 if l and not l.startswith("\\") and not l.startswith("|")]
        if len(lines) <= 1:
            return out
        # First data line after header
        data = lines[1].split(',')
        ra_wise = float(data[0])
        dec_wise = float(data[1])
        W1mag = float(data[2]) if data[2] else np.nan
        W2mag = float(data[3]) if data[3] else np.nan

        dra = (ra_wise - ra) * np.cos(np.deg2rad(dec)) * 3600.0
        ddec = (dec_wise - dec) * 3600.0
        sep_arcsec = np.sqrt(dra**2 + ddec**2)

        out.update({'ra_wise': ra_wise, 'dec_wise': dec_wise,
                    'W1mag': W1mag, 'W2mag': W2mag, 'sep_arcsec': sep_arcsec})

    except Exception as e:
        print(f"WISE query failed for RA={ra:.3f}, Dec={dec:.3f}: {e}")

    return out


# ------------------------------------------------------------------
# Query all FXRT candidates
# ------------------------------------------------------------------
matches_wise = []
for i, row in fxrt.iterrows():
    m = query_wise(row.RA_CORR, row.DEC_CORR, 2*row.R95_arcsec)
    matches_wise.append(m)
    if i % 10 == 0:
        print(f"Queried {i+1}/{len(fxrt)} sources")

wise_df = pd.DataFrame(matches_wise)
fxrt_wise = pd.concat([fxrt.reset_index(drop=True),
                       wise_df.reset_index(drop=True)], axis=1)

# ------------------------------------------------------------------
# Save merged result
# ------------------------------------------------------------------
fxrt_wise.to_csv("crossmatches/fxrt_crossmatch_wise_3sig.txt",
                 sep="\t", index=False, float_format="%.6f")

print(f"\n✅ WISE crossmatch complete: "
      f"{np.isfinite(fxrt_wise['ra_wise']).sum()} matches out of {len(fxrt_wise)} candidates.")
print("Saved to crossmatches/fxrt_crossmatch_wise_3sig.txt")

# ------------------------------------------------------------------
# Preview a few results
# ------------------------------------------------------------------
cols = ["ObsId","RA_CORR","DEC_CORR","R3sigma_arcsec",
        "ra_wise","dec_wise","W1mag","W2mag","sep_arcsec"]
fxrt_wise[cols].head(10)

```

```python
# ------------------------------------------------------------------
# GALEX (UV) crossmatch  — VizieR cone search
# Catalog: II/312/ais  (GALEX All-Sky Imaging Survey)
# ------------------------------------------------------------------



def query_galex(ra, dec, rad_arcsec=3.0, retries=3):
    """
    Query GALEX GR6/GR7 (AIS) via VizieR cone search.
    Includes retries and longer timeout to reduce SSL handshake timeouts.
    """
    base = "https://vizier.cds.unistra.fr/viz-bin/asu-tsv"
    rad_arcmin = rad_arcsec / 60.0
    params = {
        "-source": "II/312/ais",
        "-c": f"{ra} {dec}",
        "-c.r": f"{rad_arcmin:.4f}",
        "-out": "RAJ2000,DEJ2000,NUVmag,FUVmag",
        "-out.max": 1
    }
    url = base + "?" + urllib.parse.urlencode(params)
    out = {'ra_galex': np.nan, 'dec_galex': np.nan,
           'NUVmag': np.nan, 'FUVmag': np.nan, 'sep_arcsec': np.nan}

    for attempt in range(retries):
        try:
            with urllib.request.urlopen(url, timeout=120) as r:
                text = r.read().decode("utf-8")
            lines = [l for l in text.splitlines()
                     if l and not l.startswith("#") and not l.startswith("!")]
            if len(lines) <= 3:
                return out
            vals = lines[-1].split('\t')
            if len(vals) < 4:
                return out

            ra_galex = float(vals[0])
            dec_galex = float(vals[1])
            NUVmag = float(vals[2]) if vals[2] else np.nan
            FUVmag = float(vals[3]) if vals[3] else np.nan

            dra = (ra_galex - ra) * np.cos(np.deg2rad(dec)) * 3600.0
            ddec = (dec_galex - dec) * 3600.0
            sep_arcsec = np.sqrt(dra**2 + ddec**2)

            out.update({'ra_galex': ra_galex, 'dec_galex': dec_galex,
                        'NUVmag': NUVmag, 'FUVmag': FUVmag, 'sep_arcsec': sep_arcsec})
            return out

        except Exception as e:
            print(f"GALEX query attempt {attempt+1}/{retries} failed for RA={ra:.3f}, Dec={dec:.3f}: {e}")
            time.sleep(5)  # wait before retry

    # If all retries fail, return NaNs
    return out


# ------------------------------------------------------------------
# Query all FXRT candidates
# ------------------------------------------------------------------
matches_galex = []
for i, row in fxrt.iterrows():
    m = query_galex(row.RA_CORR, row.DEC_CORR, 1.405*row.R95_arcsec)
    matches_galex.append(m)
    if i % 10 == 0:
        print(f"Queried {i+1}/{len(fxrt)} sources")

galex_df = pd.DataFrame(matches_galex)
fxrt_galex = pd.concat([fxrt.reset_index(drop=True),
                        galex_df.reset_index(drop=True)], axis=1)

# ------------------------------------------------------------------
# Save merged result
# ------------------------------------------------------------------
fxrt_galex.to_csv("crossmatches/fxrt_crossmatch_galex_3sig.txt",
                  sep="\t", index=False, float_format="%.6f")

print(f"\n✅ GALEX crossmatch complete: "
      f"{np.isfinite(fxrt_galex['ra_galex']).sum()} matches out of {len(fxrt_galex)} candidates.")
print("Saved to crossmatches/fxrt_crossmatch_galex_3sig.txt")

# ------------------------------------------------------------------
# Preview a few results
# ------------------------------------------------------------------
cols = ["ObsId","RA_CORR","DEC_CORR","R95_arcsec",
        "ra_galex","dec_galex","NUVmag","FUVmag","sep_arcsec"]
fxrt_galex[cols].head(10)
```

```python
# ------------------------------------------------------------------
# DECaLS / DESI Legacy Imaging Survey (DR10) crossmatch
"for the covarage between -80 < b < -30"
# ------------------------------------------------------------------

def query_decals(ra, dec, rad_arcsec=3.0, retries=1):
    """
    Query DECaLS (DESI Legacy Imaging Survey DR10) catalog via NOIRLab API.
    Returns dict with RA/Dec, g/r/z magnitudes, and separation (arcsec).
    """
    # Convert radius to degrees
    rad_deg = rad_arcsec / 3600.0

    # Base URL for Legacy Survey catalog (DR10)
    base = "https://www.legacysurvey.org/viewer/cat.json"

    # Build query parameters
    params = {
        "ra": f"{ra:.6f}",
        "dec": f"{dec:.6f}",
        "layer": "dr10",
        "radius": f"{rad_deg:.6f}"
    }

    url = base + "?" + urllib.parse.urlencode(params)
    out = {'ra_decals': np.nan, 'dec_decals': np.nan,
           'gmag': np.nan, 'rmag': np.nan, 'zmag': np.nan, 'sep_arcsec': np.nan}

    for attempt in range(retries):
        try:
            with urllib.request.urlopen(url, timeout=60) as r:
                data = r.read().decode("utf-8")

            if '"sources": []' in data:
                return out

            # Parse JSON manually (no json import needed)
            # Extract the first source block
            src_start = data.find('"sources": [') + len('"sources": [')
            src_end = data.find(']', src_start)
            src_block = data[src_start:src_end]

            def extract(field):
                key = f'"{field}": '
                start = src_block.find(key)
                if start == -1:
                    return np.nan
                val_start = start + len(key)
                val_end = src_block.find(',', val_start)
                val = src_block[val_start:val_end].strip()
                try:
                    return float(val)
                except:
                    return np.nan

            ra_d = extract("ra")
            dec_d = extract("dec")
            gmag = extract("decam_mag_g")
            rmag = extract("decam_mag_r")
            zmag = extract("decam_mag_z")

            dra = (ra_d - ra) * np.cos(np.deg2rad(dec)) * 3600.0
            ddec = (dec_d - dec) * 3600.0
            sep_arcsec = np.sqrt(dra**2 + ddec**2)

            out.update({'ra_decals': ra_d, 'dec_decals': dec_d,
                        'gmag': gmag, 'rmag': rmag, 'zmag': zmag,
                        'sep_arcsec': sep_arcsec})
            return out

        except Exception as e:
            print(f"DECaLS query attempt {attempt+1}/{retries} failed for RA={ra:.3f}, Dec={dec:.3f}: {e}")
            time.sleep(5)

    return out


# ------------------------------------------------------------------
# Query all FXRT candidates
# ------------------------------------------------------------------
matches_decals = []
for i, row in fxrt.iterrows():
    # Only query for sources south of PS1 footprint
    if row.DEC_CORR < -30.0:
        m = query_decals(row.RA_CORR, row.DEC_CORR, 1.405*row.R95_arcsec)
    else:
        m = {'ra_decals': np.nan, 'dec_decals': np.nan,
             'gmag': np.nan, 'rmag': np.nan, 'zmag': np.nan, 'sep_arcsec': np.nan}
    matches_decals.append(m)
    if i % 10 == 0:
        print(f"Queried {i+1}/{len(fxrt)} sources")

decals_df = pd.DataFrame(matches_decals)
fxrt_decals = pd.concat([fxrt.reset_index(drop=True),
                         decals_df.reset_index(drop=True)], axis=1)

# ------------------------------------------------------------------
# Save merged result
# ------------------------------------------------------------------
fxrt_decals.to_csv("crossmatches/fxrt_crossmatch_decals_3sig.txt",
                   sep="\t", index=False, float_format="%.6f")

print(f"\n✅ DECaLS crossmatch complete: "
      f"{np.isfinite(fxrt_decals['ra_decals']).sum()} matches out of {len(fxrt_decals)} candidates.")
print("Saved to crossmatches/fxrt_crossmatch_decals_3sig.txt")

# ------------------------------------------------------------------
# Preview a few results
# ------------------------------------------------------------------
cols = ["ObsId","RA_CORR","DEC_CORR","R95_arcsec",
        "ra_decals","dec_decals","gmag","rmag","zmag","sep_arcsec"]
fxrt_decals[cols].head(10)
```
