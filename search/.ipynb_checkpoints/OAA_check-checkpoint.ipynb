```python
import subprocess
import re

def get_oaa(evtfile, ra, dec):
    cmd = [
        "dmcoords",
        evtfile,
        "opt=cel",
        f"ra={ra}",
        f"dec={dec}"
    ]
    
    # run CIAO command
    result = subprocess.run(cmd, capture_output=True, text=True)
    output = result.stdout
    
    # parse the "off_axis_angle"
    match = re.search(r"off_axis_angle:\s*([\d\.]+)", output)
    if match:
        return float(match.group(1))
    else:
        raise ValueError("Could not parse off-axis angle from dmcoords output.")

```

```python
evt = "detections/30892/578_evt2_filtered.fits"
oaa = get_oaa(evt, ra=150.1142, dec=2.2050)
print(oaa, "arcmin")
```
