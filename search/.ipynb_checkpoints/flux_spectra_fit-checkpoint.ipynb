```python
# Manual spectrum extraction in conda CIAO environment
import os
from ciao_contrib.runtool import dmcopy, dmextract, dmstat
import numpy as np
import pandas as pd
import subprocess

# -------------------------------------------------------------------
# 1. Input definitions
# -------------------------------------------------------------------
obsid = "30892"

evtfile = f"/data/tvlieshout/fxrt/search/detections/{obsid}/578_evt2_filtered.fits"
src_reg = f"/data/tvlieshout/fxrt/search/detections/regions/source_{obsid}.reg"
bkg_reg = f"/data/tvlieshout/fxrt/search/detections/regions/bkg_{obsid}.reg"
outdir  = f"/data/tvlieshout/fxrt/search/detections/flare_fits/{obsid}"
os.makedirs(outdir, exist_ok=True)

src_pi = f"{outdir}/flare_{obsid}_src.pi"
bkg_pi = f"{outdir}/flare_{obsid}_bkg.pi"

# -------------------------------------------------------------------
# 2. Extract source & background PHA spectra
# -------------------------------------------------------------------
print("üîß Extracting source spectrum ...")
dmextract.punlearn()
dmextract.infile  = f"{evtfile}[sky=region({src_reg})][bin pi]"
dmextract.outfile = src_pi
dmextract.clobber = "yes"
dmextract()

print("üîß Extracting background spectrum ...")
dmextract.punlearn()
dmextract.infile  = f"{evtfile}[sky=region({bkg_reg})][bin pi]"
dmextract.outfile = bkg_pi
dmextract.clobber = "yes"
dmextract()

# -------------------------------------------------------------------
# 3. Measure total counts & exposure
# -------------------------------------------------------------------
# -------------------------------------------------------------------
# 3. Measure total counts & exposure
# -------------------------------------------------------------------


def get_counts(file):
    """Sum the COUNTS column in a PHA file."""
    out = subprocess.check_output(["dmlist", f"{file}[cols counts]", "data"], text=True)
    total = 0
    for line in out.splitlines():
        parts = line.strip().split()
        if len(parts) == 2 and parts[0].isdigit():
            total += float(parts[1])
    return total


src_counts = get_counts(src_pi)
bkg_counts = get_counts(bkg_pi)

# exposure block
import pandas as pd
import numpy as np

# Path to your photon CSV (adjust if needed)
photon_csv = f"detections/correct_photon_tables/photons_{obsid}.csv"

# Read photon arrival times
df = pd.read_csv(photon_csv)
if "Time_s_since_obs_start" not in df.columns:
    raise KeyError("Column 'Time_s_since_obs_start' not found in photon CSV!")

tmin, tmax = df["Time_s_since_obs_start"].min(), df["Time_s_since_obs_start"].max()
expo = tmax - tmin
print(f"üïí Computed exposure from photon CSV: {expo:.1f} s")



# -------------------------------------------------------------------
# 4. Background‚Äìsubtracted count rate and approximate flux
# -------------------------------------------------------------------
net_counts = src_counts - bkg_counts
count_rate = net_counts / expo

A_eff = 400.0          # cm^2 (typical ACIS)
E_mean_keV = 1.5       # mean photon energy
flux_ergcm2s = count_rate * E_mean_keV * 1.602e-9 / A_eff
fluence_ergcm2 = flux_ergcm2s * expo

print(f"‚ö° Net count rate ‚âà {count_rate:.4f} cts/s")
print(f"‚ö° Energy flux ‚âà {flux_ergcm2s:.2e} erg/cm¬≤/s")
print(f"‚ö° Fluence     ‚âà {fluence_ergcm2:.2e} erg/cm¬≤")

# -------------------------------------------------------------------
# 5. Save quick summary
# -------------------------------------------------------------------
summary = {
    "ObsID": {obsid},
    "src_counts": src_counts,
    "bkg_counts": bkg_counts,
    "net_counts": net_counts,
    "exposure_s": expo,
    "count_rate": count_rate,
    "flux_ergcm2s": flux_ergcm2s,
    "fluence_ergcm2": fluence_ergcm2
}
pd.DataFrame([summary]).to_csv(f"{outdir}/manual_flux_summary_{obsid}.csv", index=False)
print(f"üìÑ Saved summary ‚Üí {outdir}/manual_flux_summary_{obsid}.csv")
```

```python
import os
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ciao_contrib.runtool import dmcopy, specextract
from sherpa.astro.ui import *

# ============================================================
# 1Ô∏è‚É£ Galactic NH per ObsID
# ============================================================
nh_cm2_by_obsid = {
    "30892": 1.92e20,
    "28964": 7.76e20,
    "29128": 2.95e20,
    "18927": 2.05e21
}

# ============================================================
# 2Ô∏è‚É£ Helper functions
# ============================================================
def get_total_counts(pi_file):
    out = subprocess.check_output(["dmlist", f"{pi_file}[cols counts]", "data"], text=True)
    total = 0
    for line in out.splitlines():
        parts = line.strip().split()
        if len(parts) == 2 and parts[0].isdigit():
            total += float(parts[1])
    return total

def read_key(pi_file, key):
    out = subprocess.check_output(["dmkeypar", pi_file, key, "echo=yes"], text=True)
    return float(out.strip())

# ============================================================
# 3Ô∏è‚É£ Loop through ObsIDs
# ============================================================
results = []

for obsid, nh_val in nh_cm2_by_obsid.items():
    print(f"\n=== Processing ObsID {obsid} ===")

    evtfile = f"detections/{obsid}/acisf{obsid}N001_evt2.fits"
    srcreg = f"detections/regions/source_{obsid}.reg"
    bkgreg = f"detections/regions/bkg_{obsid}.reg"
    photon_csv = f"detections/correct_photon_tables/photons_{obsid}.csv"
    outroot = f"detections/fluxfit/{obsid}/{obsid}_flare"
    os.makedirs(os.path.dirname(outroot), exist_ok=True)

    # Load photon times
    df = pd.read_csv(photon_csv)
    times = df["Time_s_since_obs_start"].values
    energies = df["Energy_keV"].values
    tmin, tmax = times.min(), times.max()
    exposure = tmax - tmin
    t90 = np.percentile(times, 95) - np.percentile(times, 5)
    print(f"Flare window: {tmin:.1f}‚Äì{tmax:.1f} s (Œît={exposure:.1f} s, T90={t90:.1f} s)")

    # Filter event file
    flare_evt = f"detections/{obsid}/flare_evt2.fits"
    dmcopy(f"{evtfile}[time={tmin}:{tmax}]", flare_evt, clobber=True)

    # Extract spectra
    specextract.punlearn()
    specextract.infile = f"{flare_evt}[sky=region({srcreg})]"
    specextract.bkgfile = f"{flare_evt}[sky=region({bkgreg})]"
    specextract.outroot = outroot
    specextract.correctpsf = True
    specextract.weight = False
    specextract.clobber = True
    specextract()

    src_pi = f"{outroot}.pi"
    bkg_pi = f"{outroot}_bkg.pi"

    # Counts & scaling
    Csrc = get_total_counts(src_pi)
    Cbkg = get_total_counts(bkg_pi)
    BACKSCAL_src = read_key(src_pi, "BACKSCAL")
    BACKSCAL_bkg = read_key(bkg_pi, "BACKSCAL")
    Tsrc = read_key(src_pi, "EXPOSURE")
    Tbkg = read_key(bkg_pi, "EXPOSURE")
    alpha = (BACKSCAL_src / BACKSCAL_bkg) * (Tsrc / Tbkg)
    net_counts = Csrc - alpha * Cbkg
    count_rate = net_counts / exposure
    print(f"Counts: src={Csrc:.1f}, bkg={Cbkg:.1f}, net={net_counts:.1f}, rate={count_rate:.4f} cts/s")

    # ============================================================
    # 4Ô∏è‚É£ Adaptive binning for 3 counts/bin light curve
    # ============================================================
    min_counts_per_bin = 3
    times_sorted = np.sort(times)
    bins = [times_sorted[0]]
    count = 0
    for t in times_sorted:
        count += 1
        if count >= min_counts_per_bin:
            bins.append(t)
            count = 0
    if bins[-1] != times_sorted[-1]:
        bins.append(times_sorted[-1])

    counts, edges = np.histogram(times_sorted, bins=bins)
    centers = 0.5 * (edges[:-1] + edges[1:])
    bin_durations = np.diff(edges)
    rate = counts / bin_durations
    rate_err = np.sqrt(counts) / bin_durations

    # ============================================================
    # 5Ô∏è‚É£ Sherpa model fits
    # ============================================================
    nh_1e22 = nh_val / 1e22
    clean()
    load_pha(src_pi)
    group_counts(3)
    set_stat("wstat")
    set_xsabund("wilm")
    set_xsxsect("vern")
    ignore(None)
    notice(0.3, 10.0)

    # ---------- Power Law ----------
    set_source(xsphabs.abs1 * xspowerlaw.pl)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    fit()
    gamma_pl = pl.PhoIndex.val
    f_abs_pl = calc_energy_flux(0.3, 10.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_pl = calc_energy_flux(0.3, 10.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    W_pl = get_stat_info()[0].statval
    print(f"PL  : Œì={gamma_pl:.2f}, Wstat={W_pl:.2f}")

    # ---------- Broken Power Law ----------
    set_source(xsphabs.abs1 * xsbknpower.bkn)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    fit()
    gamma1_bpl = bkn.PhoIndx1.val
    gamma2_bpl = bkn.PhoIndx2.val
    Ebreak = bkn.BreakE.val
    f_abs_bpl = calc_energy_flux(0.3, 10.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_bpl = calc_energy_flux(0.3, 10.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    W_bpl = get_stat_info()[0].statval
    print(f"BPL : Œì‚ÇÅ={gamma1_bpl:.2f}, Œì‚ÇÇ={gamma2_bpl:.2f}, Eb={Ebreak:.2f} keV, W={W_bpl:.2f}")

    # ---------- Blackbody ----------
    set_source(xsphabs.abs1 * xsbbody.bb)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    fit()
    kT_bb = bb.kT.val
    f_abs_bb = calc_energy_flux(0.3, 10.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_bb = calc_energy_flux(0.3, 10.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    W_bb = get_stat_info()[0].statval
    print(f"BB  : kT={kT_bb:.2f} keV, W={W_bb:.2f}")

    # ---------- Bremsstrahlung ----------
    set_source(xsphabs.abs1 * xsbremss.brem)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    fit()
    kT_brem = brem.kT.val
    f_abs_brem = calc_energy_flux(0.3, 10.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_brem = calc_energy_flux(0.3, 10.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    W_brem = get_stat_info()[0].statval
    print(f"BREM: kT={kT_brem:.2f} keV, W={W_brem:.2f}")

    # ============================================================
    # 6Ô∏è‚É£ Plot 1.1: Binned light curve (time vs flux) + all models with avg flux 
    # ============================================================

    plt.figure(figsize=(7, 4))
    
    # Compute flux per time bin using unabsorbed model flux (simplified)
    # You already have photon times and bins from before:
    min_counts_per_bin = 3
    times_sorted = np.sort(times)
    bins = [times_sorted[0]]
    count = 0
    for t in times_sorted:
        count += 1
        if count >= min_counts_per_bin:
            bins.append(t)
            count = 0
    if bins[-1] != times_sorted[-1]:
        bins.append(times_sorted[-1])
    counts, edges = np.histogram(times_sorted, bins=bins)
    centers = 0.5 * (edges[:-1] + edges[1:])
    bin_durations = np.diff(edges)

    # Convert counts to count rate (cts/s)
    count_rate_bin = counts / bin_durations
    
    # Estimate flux scaling from the *total* unabsorbed PL flux
    flux_scale = f_unabs_pl / count_rate if count_rate > 0 else 1
    flux_bin = count_rate_bin * flux_scale
    
    plt.errorbar(centers, flux_bin, 
                 xerr=0.5*bin_durations, 
                 yerr=np.sqrt(counts)/bin_durations*flux_scale,
                 fmt='o', color='black', label='Data (3 counts/bin)')
    
    # Overlay horizontal model fluxes (for reference)
    plt.axhline(f_unabs_pl, color='crimson', lw=2, label='PL Model (avg)')
    plt.axhline(f_unabs_bpl, color='royalblue', lw=2, label='BPL Model (avg)')
    plt.axhline(f_unabs_bb, color='green', lw=2, label='BB Model (avg)')
    plt.axhline(f_unabs_brem, color='orange', lw=2, label='BREMSS Model (avg)')

    plt.xscale('linear')
    plt.yscale('log')
    plt.xlabel("Time since obs start (s)")
    plt.ylabel("Flux (erg cm‚Åª¬≤ s‚Åª¬π)")
    plt.title(f"ObsID {obsid} ‚Äî Time vs Average Flux Lightcurve")
    plt.legend(frameon=False)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{outroot}_lightcurve_avg_flux.png", dpi=150, bbox_inches='tight')
    plt.show()

    # -----------------------------------------------------
    # 6Ô∏è‚É£ Plot 1.2 light curve with adaptive 3-count bins + 4 model flux conversions
    # -----------------------------------------------------
    
    min_counts_per_bin = 3
    
    # Sort photon times
    times_sorted = np.sort(times)
    bins = [times_sorted[0]]
    count = 0
    
    # Build variable-width bins with at least 3 counts each
    for t in times_sorted:
        count += 1
        if count >= min_counts_per_bin:
            bins.append(t)
            count = 0
    if bins[-1] != times_sorted[-1]:
        bins.append(times_sorted[-1])

    # Bin data
    counts, edges = np.histogram(times_sorted, bins=bins)
    centers = 0.5 * (edges[:-1] + edges[1:])
    bin_durations = np.diff(edges)
    t_err = 0.5 * bin_durations  # half-width of each time bin (for x-error bars)
    
    # Convert to count rate and Poisson errors
    rate = counts / bin_durations
    rate_err = np.sqrt(counts) / bin_durations
    
    # -----------------------------------------------------
    # üîπ Convert to flux using best-fit model conversion factors
    # -----------------------------------------------------
    # Choose absorbed or unabsorbed fluxes depending on your preference
    # (here: unabsorbed to compare intrinsic emission)
    CF_pl   = f_unabs_pl   / count_rate
    CF_bpl  = f_unabs_bpl  / count_rate
    CF_bb   = f_unabs_bb   / count_rate
    CF_brem = f_unabs_brem / count_rate
    
    flux_pl_t   = rate * CF_pl
    flux_bpl_t  = rate * CF_bpl
    flux_bb_t   = rate * CF_bb
    flux_brem_t = rate * CF_brem
    
    flux_err_t = rate_err * CF_pl  # use PL scaling for y-errors

    # -----------------------------------------------------
    # üîπ Plot light curve with all model-scaled fluxes
    # -----------------------------------------------------
    plt.figure(figsize=(7, 4))
    plt.errorbar(centers - tmin, flux_pl_t, xerr=t_err, yerr=flux_err_t,
                 fmt='o', color='black', ecolor='gray', ms=4,
                 label='Observed flux (‚â•3 counts/bin)')
    
    plt.plot(centers - tmin, flux_pl_t,   color='crimson', lw=2, label='PL-scaled flux')
    plt.plot(centers - tmin, flux_bpl_t,  color='royalblue', lw=2, label='BPL-scaled flux')
    plt.plot(centers - tmin, flux_bb_t,   color='green', lw=2, label='BB-scaled flux')
    plt.plot(centers - tmin, flux_brem_t, color='orange', lw=2, label='BREMSS-scaled flux')
    
    plt.xlabel("Time since flare start (s)")
    plt.ylabel("Flux (erg cm$^{-2}$ s$^{-1}$)")
    plt.title(f"ObsID {obsid} ‚Äî Time vs Flux lightcurve  ")
    plt.xscale('log')
    plt.yscale('log')
    plt.legend(frameon=False, fontsize=8)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{outroot}_lightcurve_allmodels_3ctbin.png", dpi=150, bbox_inches='tight')
    plt.show()



    # ============================================================
    # 7Ô∏è‚É£ Plot 2: Spectrum (Energy vs Counts/s/keV)
    # ============================================================
    load_pha(src_pi)
    group_counts(3)
    ignore(None); notice(0.3, 10.0)
    plot_data(clearwindow=True)
    pdata = get_data_plot()
    E = pdata.x
    C = pdata.y
    Cerr = pdata.yerr
    E_err = 0.5 * (pdata.xhi - pdata.xlo)

    plt.figure(figsize=(7, 4))
    plt.errorbar(E, C, xerr=E_err, yerr=Cerr, fmt='o', ms=3, color='black', label='Data (3 counts/bin)')

    # overlay all models
    for model, color, label in [
        ("xspowerlaw.pl", "crimson", "PL"),
        ("xsbknpower.bkn", "royalblue", "BPL"),
        ("xsbbody.bb", "green", "BB"),
        ("xsbremss.brem", "orange", "BREMSS")
    ]:
        set_source(xsphabs.abs1 * eval(model.split('.')[-1]))
        abs1.nh = nh_1e22; freeze(abs1.nh)
        
        m = get_model_plot()
        plt.plot(m.x, m.y, color=color, lw=2, label=label)

    plt.xscale('log'); plt.yscale('log')
    plt.xlabel("Energy (keV)")
    plt.ylabel("Counts s‚Åª¬π keV‚Åª¬π")
    plt.title(f"ObsID {obsid} ‚Äî Spectral Fits (3 counts/bin)")
    plt.legend(frameon=False)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"{outroot}_spectrum_allmodels.png", dpi=150, bbox_inches='tight')
    plt.show()

    # ============================================================
    # 8Ô∏è‚É£ Plot 3: Hardness ratio evolution
    # ============================================================
    soft_band = (0.3, 2.0)
    hard_band = (2.0, 10.0)
    soft_counts, hard_counts = [], []
    t_centers = 0.5 * (edges[:-1] + edges[1:])
    t_err = 0.5 * (edges[1:] - edges[:-1])
    for lo, hi in zip(edges[:-1], edges[1:]):
        mask = (times >= lo) & (times < hi)
        ebin = energies[mask]
        soft_counts.append(np.sum((ebin >= soft_band[0]) & (ebin < soft_band[1])))
        hard_counts.append(np.sum((ebin >= hard_band[0]) & (ebin <= hard_band[1])))
    soft_counts, hard_counts = np.array(soft_counts), np.array(hard_counts)
    HR = (hard_counts - soft_counts) / (hard_counts + soft_counts + 1e-6)
    dH, dS = np.sqrt(hard_counts), np.sqrt(soft_counts)
    dHR = 2 / ((hard_counts + soft_counts)**2 + 1e-6) * np.sqrt(
        (soft_counts**2)*(dH**2) + (hard_counts**2)*(dS**2)
    )

    plt.figure(figsize=(7, 4))
    plt.errorbar(t_centers, HR, xerr=t_err, yerr=dHR, fmt='o', color='black',
                 ecolor='gray', capsize=2, label='Hardness Ratio')
    plt.axhline(0, color='red', lw=1, ls='--')
    plt.xlabel("Time since obs start (s)")
    plt.ylabel("Hardness Ratio (H‚àíS)/(H+S)")
    plt.title(f"ObsID {obsid} ‚Äî HR Evolution (3 counts/bin)")
    plt.grid(alpha=0.3)
    plt.legend(frameon=False)
    plt.tight_layout()
    plt.savefig(f"{outroot}_HR_evolution.png", dpi=150, bbox_inches='tight')
    plt.show()

    # ============================================================
    # 9Ô∏è‚É£ Store all model fit results
    # ============================================================
    
    results.append({
        "ObsID": obsid,
        "Exposure_s": exposure,
        "Net_counts": net_counts,
        "Count_rate": count_rate,
        
        # Power law
        "Gamma_PL": gamma_pl,
        "Flux_abs_PL": f_abs_pl,
        "Flux_unabs_PL": f_unabs_pl,
        "W_PL": W_pl,
        
        # Broken power law
        "Gamma1_BPL": gamma1_bpl,
        "Gamma2_BPL": gamma2_bpl,
        "Ebreak_keV": Ebreak,
        "Flux_abs_BPL": f_abs_bpl,
        "Flux_unabs_BPL": f_unabs_bpl,
        "W_BPL": W_bpl,
        
        # Blackbody
        "kT_BB": kT_bb,
        "Flux_abs_BB": f_abs_bb,
        "Flux_unabs_BB": f_unabs_bb,
        "W_BB": W_bb,
        
        # Bremsstrahlung
        "kT_BREM": kT_brem,
        "Flux_abs_BREM": f_abs_brem,
        "Flux_unabs_BREM": f_unabs_brem,
        "W_BREM": W_brem,
    })

    # -----------------------------------------------------
    # üîπ Compute number of spectral bins used (for AIC/BIC)
    # -----------------------------------------------------
    try:
        # After fitting, Sherpa stores data in dataset 1
        d = get_data()
        n_bins = len(d.y)  # number of data points actually used
    except Exception:
        n_bins = np.nan  # fallback if not available
    
    # Store it along with model W-stats so we can use later
    results[-1]["n_bins"] = n_bins


# ============================================================
# üîü Save and summarize results
# ============================================================
df = pd.DataFrame(results)
summary_path = "detections/fluxfit/flare_flux_summary_PL_BPL_BB_BREM.csv"
df.to_csv(summary_path, index=False)
print(f"\nüìÑ Saved summary ‚Üí {summary_path}")

# ---- Compute best-fitting model per ObsID ----
fit_cols = ["W_PL", "W_BPL", "W_BB", "W_BREM"]
df["Best_Model"] = df[fit_cols].idxmin(axis=1).str.replace("W_", "")

# ---- Compute ŒîW (difference from best model) ----
df["ŒîW_to_best"] = df[fit_cols].min(axis=1)
for col in fit_cols:
    df[f"ŒîW_vs_{col.split('_')[-1]}"] = df[col] - df["ŒîW_to_best"]

# Compute AIC and BIC for each model automatically
k_dict = {"PL": 2, "BPL": 4, "BB": 2, "BREM": 2}

for model in ["PL", "BPL", "BB", "BREM"]:
    k = k_dict[model]
    df[f"AIC_{model}"] = 2 * k + df[f"W_{model}"]
    df[f"BIC_{model}"] = k * np.log(df["n_bins"]) + df[f"W_{model}"]

# Identify the best model by each criterion
df["Best_Model_AIC"] = df[[f"AIC_{m}" for m in k_dict]].idxmin(axis=1).str.replace("AIC_", "")
df["Best_Model_BIC"] = df[[f"BIC_{m}" for m in k_dict]].idxmin(axis=1).str.replace("BIC_", "")


# ---- Display compact summary ----
summary_cols = [
    "ObsID", "n_bins", "Best_Model", "Best_Model_AIC", "Best_Model_BIC",
    "W_PL", "W_BPL", "W_BB", "W_BREM",
    "AIC_PL", "AIC_BPL", "AIC_BB", "AIC_BREM",
    "BIC_PL", "BIC_BPL", "BIC_BB", "BIC_BREM",
    "Flux_unabs_PL", "Flux_unabs_BPL", "Flux_unabs_BB", "Flux_unabs_BREM"
]
display(df[summary_cols].round(3))


# ---- Pretty color highlighting (extended for AIC/BIC) ----
def highlight_best_models(row):
    styles = [""] * len(row)

    # Colors for each selection criterion
    colors = {
        "W": "background-color: lightgreen",    # raw W-stat
        "AIC": "background-color: lightblue",   # AIC best
        "BIC": "background-color: pink"         # BIC best
    }

    # Highlight best W
    best_col_W = f"W_{row['Best_Model']}"
    if best_col_W in row.index:
        styles[row.index.get_loc(best_col_W)] = colors["W"]

    # Highlight best AIC
    best_col_AIC = f"AIC_{row['Best_Model_AIC']}"
    if best_col_AIC in row.index:
        styles[row.index.get_loc(best_col_AIC)] = colors["AIC"]

    # Highlight best BIC
    best_col_BIC = f"BIC_{row['Best_Model_BIC']}"
    if best_col_BIC in row.index:
        styles[row.index.get_loc(best_col_BIC)] = colors["BIC"]

    return styles

display(df.style.apply(highlight_best_models, axis=1))


# ---- Print quick text summary ----
for _, row in df.iterrows():
    w_best = row["Best_Model"]
    aic_best = row["Best_Model_AIC"]
    bic_best = row["Best_Model_BIC"]
    w_val = row[f"W_{w_best}"]
    print(f"ObsID {row['ObsID']}:")
    print(f"  ‚Ä¢ W-stat best ‚Üí {w_best} (W={w_val:.2f})")
    print(f"  ‚Ä¢ AIC best    ‚Üí {aic_best}")
    print(f"  ‚Ä¢ BIC best    ‚Üí {bic_best}")
    print()

```

```python
import os
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ciao_contrib.runtool import dmcopy, specextract
from sherpa.astro.ui import *
from scipy.stats import beta


# ============================================================
# 1Ô∏è‚É£ Galactic NH per ObsID
# ============================================================
nh_cm2_by_obsid = {
    "30892": 1.92e20,
    "28964": 7.76e20
}

# ============================================================
# 2Ô∏è‚É£ Helper functions
# ============================================================
def get_total_counts(pi_file):
    out = subprocess.check_output(["dmlist", f"{pi_file}[cols counts]", "data"], text=True)
    total = 0
    for line in out.splitlines():
        parts = line.strip().split()
        if len(parts) == 2 and parts[0].isdigit():
            total += float(parts[1])
    return total

def read_key(pi_file, key):
    out = subprocess.check_output(["dmkeypar", pi_file, key, "echo=yes"], text=True)
    return float(out.strip())




# ============================================================
# 3Ô∏è‚É£ Main loop
# ============================================================
results = []

for obsid, nh_val in nh_cm2_by_obsid.items():
    print(f"\n=== Processing ObsID {obsid} ===")

    evtfile = f"detections/{obsid}/acisf{obsid}N001_evt2.fits"
    srcreg = f"detections/regions/source_{obsid}.reg"
    bkgreg = f"detections/regions/bkg_{obsid}.reg"
    photon_csv = f"detections/correct_photon_tables/photons_{obsid}.csv"
    outroot = f"detections/fluxfit/{obsid}/{obsid}_flare"
    os.makedirs(os.path.dirname(outroot), exist_ok=True)

    # --- Load photon times ---
    df = pd.read_csv(photon_csv)
    times = df["Time_s_since_obs_start"].values
    energies = df["Energy_keV"].values
    tmin, tmax = times.min(), times.max()
    exposure = tmax - tmin
    t90 = np.percentile(times, 95) - np.percentile(times, 5)
    print(f"Flare window: {tmin:.1f}‚Äì{tmax:.1f} s (Œît={exposure:.1f} s, T90={t90:.1f} s)")

    # ============================================================
    # TOTAL hardness ratio for the entire flare
    # ============================================================
    soft_band = (0.3, 2.0)
    hard_band = (2.0, 7.0)
    
    S_tot = np.sum((energies >= soft_band[0]) & (energies < soft_band[1]))
    H_tot = np.sum((energies >= hard_band[0]) & (energies <= hard_band[1]))
    
    N_tot = S_tot + H_tot
    HR_total = (H_tot - S_tot) / N_tot if N_tot > 0 else np.nan
    
    print(f"Total HR: {HR_total:.3f} (Soft={S_tot}, Hard={H_tot})")
    

    # --- Filter event file ---
    flare_evt = f"detections/{obsid}/flare_evt2.fits"
    dmcopy(f"{evtfile}[time={tmin}:{tmax}]", flare_evt, clobber=True)

    # --- Extract spectra ---
    specextract.punlearn()
    specextract.infile = f"{flare_evt}[sky=region({srcreg})]"
    specextract.bkgfile = f"{flare_evt}[sky=region({bkgreg})]"
    specextract.outroot = outroot
    specextract.correctpsf = True
    specextract.weight = False
    specextract.clobber = True
    specextract()

    src_pi = f"{outroot}.pi"
    bkg_pi = f"{outroot}_bkg.pi"

    # --- Counts & scaling ---
    Csrc = get_total_counts(src_pi)
    Cbkg = get_total_counts(bkg_pi)
    BACKSCAL_src = read_key(src_pi, "BACKSCAL")
    BACKSCAL_bkg = read_key(bkg_pi, "BACKSCAL")
    Tsrc = read_key(src_pi, "EXPOSURE")
    Tbkg = read_key(bkg_pi, "EXPOSURE")
    alpha = (BACKSCAL_src / BACKSCAL_bkg) * (Tsrc / Tbkg)
    net_counts = Csrc - alpha * Cbkg
    count_rate = net_counts / exposure
    print(f"Counts: src={Csrc:.1f}, bkg={Cbkg:.1f}, net={net_counts:.1f}, rate={count_rate:.4f} cts/s")

    # ============================================================
    # TOTAL S/N for the entire flare
    # ============================================================
    sn_total = net_counts / np.sqrt(Csrc + (alpha**2) * Cbkg)
    print(f"Total S/N: {sn_total:.2f}")


    # ============================================================
    # 4Ô∏è‚É£ Sherpa model fits
    # ============================================================
    nh_1e22 = nh_val / 1e22
    clean()
    load_pha(src_pi)
    group_counts(3)
    set_stat("wstat")
    set_xsabund("wilm")
    set_xsxsect("vern")
    ignore(None)
    notice(0.3, 7.0)

    # --- Power Law ---
    set_source(xsphabs.abs1 * xspowerlaw.pl)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    
    fit()
    gamma_pl = pl.PhoIndex.val
    
    # Fit results: W-stat and Q-value (goodness)
    fr_pl = get_fit_results()
    W_pl  = fr_pl.statval
    g_pl  = fr_pl.qval
    
    f_abs_pl = calc_energy_flux(0.3, 7.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_pl = calc_energy_flux(0.3, 7.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    
    print(f"PL   : Œì={gamma_pl:.2f}, W={W_pl:.2f}, Q={g_pl:.3f}")


    # --- Blackbody ---
    set_source(xsphabs.abs1 * xsbbody.bb)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    
    fit()
    kT_bb = bb.kT.val
    
    fr_bb = get_fit_results()
    W_bb  = fr_bb.statval
    g_bb  = fr_bb.qval
    
    f_abs_bb = calc_energy_flux(0.3, 7.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_bb = calc_energy_flux(0.3, 7.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    
    print(f"BB   : kT={kT_bb:.2f} keV, W={W_bb:.2f}, Q={g_bb:.3f}")


    # --- Bremsstrahlung ---
    set_source(xsphabs.abs1 * xsbremss.brem)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    
    fit()
    kT_brem = brem.kT.val
    
    fr_brem = get_fit_results()
    W_brem  = fr_brem.statval
    g_brem  = fr_brem.qval
    
    f_abs_brem = calc_energy_flux(0.3, 7.0)
    abs1.nh = 0.0; freeze(abs1.nh)
    f_unabs_brem = calc_energy_flux(0.3, 7.0)
    abs1.nh = nh_1e22; freeze(abs1.nh)
    
    print(f"BREM : kT={kT_brem:.2f} keV, W={W_brem:.2f}, Q={g_brem:.3f}")


    # ============================================================
    # Adaptive n-count light curve + model-scaled fluxes
    # ============================================================
    minimal_counts = 3
    min_counts_per_bin = minimal_counts
    times_sorted = np.sort(times)
    bins = [times_sorted[0]]
    count = 0
    for t in times_sorted:
        count += 1
        if count >= min_counts_per_bin:
            bins.append(t)
            count = 0
    if bins[-1] != times_sorted[-1]:
        bins.append(times_sorted[-1])

    counts, edges = np.histogram(times_sorted, bins=bins)
    centers = 0.5 * (edges[:-1] + edges[1:])
    bin_durations = np.diff(edges)
    t_err = 0.5 * bin_durations
    rate = counts / bin_durations
    rate_err = np.sqrt(counts) / bin_durations

    CF_pl = f_unabs_pl / count_rate
    CF_bb = f_unabs_bb / count_rate
    CF_brem = f_unabs_brem / count_rate

    flux_pl_t = rate * CF_pl
    flux_bb_t = rate * CF_bb
    flux_brem_t = rate * CF_brem
    flux_err_t = rate_err * CF_pl

    # ============================================================
    # Spectrum (Energy vs Counts)
    # ============================================================
    load_pha(src_pi)
    group_counts(minimal_counts)
    ignore(None); notice(0.3, 7.0)
    plot_data(clearwindow=True)
    pdata = get_data_plot()
    E = pdata.x
    C = pdata.y
    Cerr = pdata.yerr
    E_err = 0.5 * (pdata.xhi - pdata.xlo)


    # ============================================================
    # 2b Combined Residuals Plot ‚Äî PL, BB, BREMSS
    # ============================================================
    
    # Re-load and group the data exactly like the main spectrum plot
    load_pha(src_pi)
    group_counts(minimal_counts)
    ignore(None); notice(0.3, 7.0)
    
    # ---- Get data arrays (same as your existing code) ----
    plot_data(clearwindow=True)
    pdata = get_data_plot()
    
    E      = pdata.x
    E_err  = 0.5 * (pdata.xhi - pdata.xlo)
    data_y = pdata.y
    data_err = pdata.yerr
    
    # ---- Models to evaluate (same order & colors as your spectrum plot) ----
    model_defs = [
        ("PL",   xsphabs.abs1 * xspowerlaw.pl,  "crimson"),
        ("BB",   xsphabs.abs1 * xsbbody.bb,     "green"),
        ("BREM", xsphabs.abs1 * xsbremss.brem,  "orange")
    ]
    
    # Container for plotting
    residual_sets = []
    
    for label, model_expr, color in model_defs:

        # Assign & fit model as in your code
        set_source(model_expr)
        abs1.nh = nh_1e22; freeze(abs1.nh)
        fit()
    
        # Evaluate high-resolution model
        m = get_model_plot()
    
        # Interpolate the model onto the data bins
        model_interp = np.interp(E, m.x, m.y)
    
        # Compute residuals
        resid = (data_y - model_interp) / data_err
    
        residual_sets.append((label, resid, color))


    # ============================================================
    # 7Ô∏è‚É£ Plot 3: Hardness Ratio evolution
    # ============================================================
    soft_band = (0.3, 2.0)
    hard_band = (2.0, 7.0)
    soft_counts, hard_counts = [], []
    t_centers = 0.5 * (edges[:-1] + edges[1:])
    for lo, hi in zip(edges[:-1], edges[1:]):
        mask = (times >= lo) & (times < hi)
        ebin = energies[mask]
        soft_counts.append(np.sum((ebin >= soft_band[0]) & (ebin < soft_band[1])))
        hard_counts.append(np.sum((ebin >= hard_band[0]) & (ebin <= hard_band[1])))
    soft_counts, hard_counts = np.array(soft_counts), np.array(hard_counts)

    H = hard_counts.astype(float)
    S = soft_counts.astype(float)
    N = H + S
    
    HR = (H - S) / (N + 1e-9)
    
    mask = N > 0
    
    from scipy.stats import beta
    alpha = 0.32
    
    a_post = H[mask] + 0.5
    b_post = S[mask] + 0.5
    
    p_lo = beta.ppf(alpha/2,     a_post, b_post)
    p_hi = beta.ppf(1-alpha/2.0, a_post, b_post)
    
    HR_lo = 2*p_lo - 1
    HR_hi = 2*p_hi - 1
    
    yerr_lower = HR[mask] - HR_lo
    yerr_upper = HR_hi - HR[mask]
    
    # üîß fix negative uncertainties
    yerr_lower = np.clip(yerr_lower, 0, None)
    yerr_upper = np.clip(yerr_upper, 0, None)
    
    # assemble full-size yerr
    yerr = np.zeros((2, len(HR)))
    yerr[0, mask] = yerr_lower
    yerr[1, mask] = yerr_upper

    # ============================================================
    # Combined Plot: Flux (top) + HR (bottom)
    # ============================================================
    
    fig, (ax1, ax2) = plt.subplots(
        2, 1,
        figsize=(7, 7),
        sharex=True,
        gridspec_kw={'height_ratios': [2, 1]}
    )

    # -------- TOP: FLUX ----------
    ax1.errorbar(centers - tmin, flux_pl_t, xerr=t_err, yerr=flux_err_t,
                 fmt='o', color='black', ecolor='black', ms=4,
                 label='Observed flux (‚â•3 counts/bin)')
    
    ax1.plot(centers - tmin, flux_pl_t, color='crimson', lw=1, label='PL-scaled flux')
    ax1.plot(centers - tmin, flux_bb_t, color='green', lw=1, label='BB-scaled flux')
    ax1.plot(centers - tmin, flux_brem_t, color='orange', lw=1, label='BREMSS-scaled flux')
    
    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.set_ylabel("Unabsorbed Flux (cgs) [0.3‚Äì7 keV]")
    ax1.set_title(f"ObsID {obsid}")
    ax1.grid(alpha=0.3)

    # -------- BOTTOM: HR ----------
    ax2.errorbar(t_centers - tmin, HR, xerr=t_err, yerr=yerr,
                 fmt='o', markersize=3, color='black', ecolor='black')
    
    ax2.axhline(0, color='red', lw=1, ls='--')
    ax2.set_xscale('log')
    ax2.set_xlabel("Time since flare start (s)")
    ax2.set_ylabel("Hardness Ratio")
    ax2.grid(alpha=0.3)
    
    xmin = (centers - tmin).min()
    xmax = (centers - tmin).max()
    
    padding_factor = 0.1  # adjust to taste (0.05‚Äì0.2 are common)
    log_xmin = np.log10(xmin)
    log_xmax = np.log10(xmax)
    log_range = log_xmax - log_xmin
    
    new_log_xmin = log_xmin - padding_factor * log_range
    new_log_xmax = log_xmax + padding_factor * log_range
    
    ax1.set_xlim(10**new_log_xmin, 10**new_log_xmax)

    plt.subplots_adjust(hspace=0.1)
    plt.savefig(f"{outroot}_combined_flux_HR_{minimal_counts}ctbin.png",
                dpi=150, bbox_inches='tight')
    plt.show()

    # ============================================================
    # Spectrum + Residuals (2-panel combined figure)
    # ============================================================
    
    # Re-load & group spectrum (same as your spectrum plot)
    load_pha(src_pi)
    group_counts(minimal_counts)
    ignore(None); notice(0.3, 7.0)
    
    # ---- Get DATA arrays (your method) ----
    plot_data(clearwindow=True)
    pdata = get_data_plot()
    
    E      = pdata.x
    E_err  = 0.5 * (pdata.xhi - pdata.xlo)
    data_y    = pdata.y
    data_err  = pdata.yerr
    
    # ---- Define models to plot (same order & colors as your spectrum plot) ----
    model_defs = [
        ("PL",   xsphabs.abs1 * xspowerlaw.pl,  "crimson"),
        ("BB",   xsphabs.abs1 * xsbbody.bb,     "green"),
        ("BREM", xsphabs.abs1 * xsbremss.brem,  "orange")
    ]
    
    # Storage for model curves + residuals
    model_curves = []
    residual_sets = []
    
    # ---- Fit each model & compute model spectrum + residuals ----
    for label, model_expr, color in model_defs:

        set_source(model_expr)
        abs1.nh = nh_1e22; freeze(abs1.nh)
        fit()
    
        # High-resolution model from your own method
        m = get_model_plot()
    
        # Interpolate model onto data bins
        model_interp = np.interp(E, m.x, m.y)
    
        # Normalized residuals
        resid = (data_y - model_interp) / data_err
    
        model_curves.append((label, m.x, m.y, color))
        residual_sets.append((label, resid, color))


    # ============================================================
    # FIGURE with 2 panels
    # ============================================================
    
    fig, (ax1, ax2) = plt.subplots(
        2, 1,
        figsize=(7, 7),
        sharex=True,
        gridspec_kw={'height_ratios': [3, 1], 'hspace': 0.05}
    )
    
    # ---------------------- TOP PANEL: Spectrum ----------------------
    ax1.errorbar(E, data_y, xerr=E_err, yerr=data_err,
                 fmt='o', ms=3, color='black', label=f'Data ({minimal_counts} ct/bin)')
    
    # Plot all 3 models
    for label, mx, my, color in model_curves:
        ax1.plot(mx, my, color=color, lw=1, label=label)

    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.set_ylabel("Counts s$^{-1}$ keV$^{-1}$")
    ax1.set_title(f"ObsID {obsid}")
    ax1.grid(alpha=0.3)
    
    # ---------------------- BOTTOM PANEL: Residuals ----------------------
    for label, resid, color in residual_sets:
        ax2.errorbar(E, resid,
                     xerr=E_err,
                     yerr=np.ones_like(resid),   # ¬±1 uncertainties
                     fmt='o', ms=3,
                     color=color, ecolor=color,
                     label=label)
    
    ax2.axhline(0, color='black', lw=1)
    ax2.set_xscale('log')
    ax2.set_xlabel("Energy (keV)")
    ax2.set_ylabel("Residuals")
    ax2.grid(alpha=0.3)
    ax2.set_ylim(-2, 3) 


    # ---------------------- SAVE ----------------------
    plt.tight_layout()
    plt.savefig(f"{outroot}_spectrum_and_residuals_ALLMODELS_{minimal_counts}ctbin.png",
                dpi=150, bbox_inches='tight')
    plt.show()




    # ============================================================
    # 8Ô∏è‚É£ Save per-ObsID results
    # ============================================================
    results.append({
        "ObsID": obsid,
        "Exposure_s": exposure,
        "Net_counts": net_counts,
        "Count_rate": count_rate,
        "n_bins": n_bins,
    
        "Gamma_PL": gamma_pl,
        "Flux_abs_PL": f_abs_pl,
        "Flux_unabs_PL": f_unabs_pl,
        "W_PL": W_pl,
        "Good_PL": g_pl,
    
        "kT_BB": kT_bb,
        "Flux_abs_BB": f_abs_bb,
        "Flux_unabs_BB": f_unabs_bb,
        "W_BB": W_bb,
        "Good_BB": g_bb,
    
        "kT_BREM": kT_brem,
        "Flux_abs_BREM": f_abs_brem,
        "Flux_unabs_BREM": f_unabs_brem,
        "W_BREM": W_brem,
        "Good_BREM": g_brem,

            # NEW additions
        "Total_Soft": S_tot,
        "Total_Hard": H_tot,
        "Total_HR": HR_total,
        "SN_Total": sn_total

    })


# ============================================================
# üîü Final summary
# ============================================================
df = pd.DataFrame(results)
summary_path = f"detections/fluxfit/flare_flux_summary_PL_BB_BREM{minimal_counts}ctbin.csv"
df.to_csv(summary_path, index=False)
print(f"\nüìÑ Saved summary ‚Üí {summary_path}")

# ---- Identify best models ----
fit_cols = ["W_PL", "W_BB", "W_BREM"]
df["Best_Model"] = df[fit_cols].idxmin(axis=1).str.replace("W_", "")

# ---- Compute AIC/BIC ----
k_dict = {"PL": 2, "BB": 2, "BREM": 2}
for model in ["PL", "BB", "BREM"]:
    k = k_dict[model]
    df[f"AIC_{model}"] = 2 * k + df[f"W_{model}"]
    df[f"BIC_{model}"] = k * np.log(df["n_bins"]) + df[f"W_{model}"]

df["Best_Model_AIC"] = df[[f"AIC_{m}" for m in k_dict]].idxmin(axis=1).str.replace("AIC_", "")
df["Best_Model_BIC"] = df[[f"BIC_{m}" for m in k_dict]].idxmin(axis=1).str.replace("BIC_", "")

# ---- Display summary ----
summary_cols = [
    "ObsID", "n_bins",
    "Best_Model", "Best_Model_AIC", "Best_Model_BIC",

    "W_PL", "W_BB", "W_BREM",
    "Good_PL", "Good_BB", "Good_BREM",

    "AIC_PL", "AIC_BB", "AIC_BREM",
    "BIC_PL", "BIC_BB", "BIC_BREM",

    "Flux_unabs_PL", "Flux_unabs_BB", "Flux_unabs_BREM"

    "Total_HR", "SN_Total"
]
# Flux columns you want in scientific notation
flux_cols = [
    "Flux_abs_PL", "Flux_unabs_PL",
    "Flux_abs_BB", "Flux_unabs_BB",
    "Flux_abs_BREM", "Flux_unabs_BREM"
]

def sci_notation(x):
    return f"{x:.3e}"

# Display summary with fluxes in scientific notation
display(
    df[summary_cols].style.format({col: sci_notation for col in flux_cols})
)


# ---- Highlight best ----
def highlight_best_models(row):
    styles = [""] * len(row)
    colors = {"W": "background-color: lightgreen",
              "AIC": "background-color: lightblue",
              "BIC": "background-color: pink"}
    for key, model_key in [("W", "Best_Model"),
                           ("AIC", "Best_Model_AIC"),
                           ("BIC", "Best_Model_BIC")]:
        best_col = f"{key}_{row[model_key]}" if key != "W" else f"W_{row[model_key]}"
        if best_col in row.index:
            styles[row.index.get_loc(best_col)] = colors[key]
    return styles


display(df.style.apply(highlight_best_models, axis=1))

# ---- Text summary ----
for _, row in df.iterrows():

    best_model = row["Best_Model"]
    best_w_col = f"W_{best_model}"
    best_w = row[best_w_col]

    print(f"ObsID {row['ObsID']}:")
    print(f"  ‚Ä¢ W-stat best ‚Üí {best_model} (W={best_w:.2f})")
    print(f"  ‚Ä¢ AIC best    ‚Üí {row['Best_Model_AIC']}")
    print(f"  ‚Ä¢ BIC best    ‚Üí {row['Best_Model_BIC']}")
    print(f"  ‚Ä¢ Goodness (PL)   ‚Üí {row['Good_PL']:.3f}")
    print(f"  ‚Ä¢ Goodness (BB)   ‚Üí {row['Good_BB']:.3f}")
    print(f"  ‚Ä¢ Goodness (BREM) ‚Üí {row['Good_BREM']:.3f}")
    print()
```
